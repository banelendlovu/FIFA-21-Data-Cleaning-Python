### Contract column

```python
fifa['Contract'].dtype
```

```python
fifa['Contract'].unique()
```

```python
for index, row in fifa.iterrows():
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
# Creating a columns that display the beginning and the end of their contracts
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date = np.nan # set to not a number since we don't know when their contracts were initialized
        end_date = np.nan
        contract_length = 0
    else:
        start_date, end_date = contract.split(' ~ ')
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = end_year - start_year
    return start_date, end_date, contract_length

# Apply the function to Contract column and create new columns
new_cols = ['Contract Start', 'Contract End', 'Contract Length(years)']
new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract')+1+i, 
                column=new_cols[i], value=new_data[i]) # location of new columns
    
```

```python
fifa.sample(3)
```

```python
# Checking if new columns are accurate
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)']].sample(5)
```

```python
# Contract Categories

def categorize_contract_status(contract):
    if contract == 'Free':
        return 'Free'
    elif 'On Loan' in contract:
        return 'On Loan'
    else:
        return 'Contract'
    
# Add contract status column
fifa.insert(fifa.columns.get_loc('Contract Length(years)')+1, 
            'Contract Status', fifa['Contract'].apply(categorize_contract_status))
fifa.sample(3)
```

```python
# Checking if new columns are accurate
fifa[['Contract', 'Contract Start', 'Contract End', 'Contract Length(years)', 'Contract Status']].sample(5)
```

### Height

```python
fifa['Height'].dtype
```

```python
fifa['Height'].unique()
```

```python
def convert_height(height):
    if 'cm' in height:
        return(height.strip('cm'))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet) * 12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
    
# Apply this function to the height column
fifa['Height'] = fifa['Height'].apply(convert_height)
fifa['Height'].unique()
```

```python
fifa = fifa.rename(columns={'Height':'Height(cm)'})
fifa.sample(1)
```

### Weight

```python
def convert_weight(weight):
    if 'kg' in weight:
        return int(weight.strip('kg'))
    else:
        pounds = int(weight.strip('lbs')) 
        return round(pounds/2.205)
    
# Apply function to the Weight column
fifa['Weight'] = fifa['Weight'].apply(convert_weight)
fifa['Weight'].unique()
```

```python
fifa = fifa.rename(columns={'Weight':'Weight(kg)'})
fifa.sample(3)
```

### Hits

Number of times the player has been searched for in the FIFA database.

```python
fifa['Hits'].dtype
```

```python
fifa['Hits'].unique()
```

```python
# Clean Hits Column
def clean_hits_column(hit):
    try:
        # If the value is a string with 'K', convert to float and multiply by 1000
        if 'K' in str(hit):
            return int(float(hit.rstrip('K')) * 1000)
        # If the value is already a float, convert to int
        elif isinstance(hit, float):
            return int(hit)
        # If the value is an integer, return as is
        elif isinstance(hit, int):
            return hit
        # If the value is 'nan', return None
        elif pd.isna(hit):
            return str(None)
        # For other cases, convert to int
        else:
            return int(hit)
    except ValueError:
        # Handle cases where the conversion fails
        print(f"Unable to convert {hit} to an integer.")
        return None

# Apply the cleaning function to each element in the array
fifa['Hits'] = fifa['Hits'].apply(clean_hits_column)
fifa['Hits'].unique()
```

```python
# change the NaN to strings

fifa['Hits'] = fifa['Hits'].fillna('None')
fifa['Hits'].unique()
```

```python
# Converting floats to strings

def new_hits(hit):
    
    new_Hit = []
    
    for hit in fifa['Hits']:
        if hit != 'None':
            new_Hit.append(int(hit))
        else:
            new_Hit.append(hit)
        return new_Hit

# Apply new_Hit to fifa
fifa['Player Hits'] = new_Hit
```
